```python
#import dependencies
import pandas as pd
import numpy as np
import requests
import json
#reference api key
from config import api_key
```

```python
#establish base url --currently using sandbox api for code exploration
base_url = "http://sandbox-api.brewerydb.com/v2"
```

```python
#look at all breweries to get id and brewery names
breweryname = []
breweryid = []
breweries_query = "breweries"
#create url to request list of breweries
#breweries_url = base_url + "/" + breweries_query +"/"+ "?key=" + api_key + "&p=1"

#loop through each page of brewery lists
for page_num in range(1):
    breweries_url = base_url + "/" + breweries_query +"/"+ "?key=" + api_key + "&p=" + str(page_num)
    print("Downloading", breweries_url)
    breweries_response = requests.get(breweries_url).json()
    #For every iteration above, loop through each brewery on the page
    for brewery in breweries_response['data']:
        breweryname.append(brewery['name'])
        breweryid.append(brewery['id'])
        print(brewery['name'], brewery['id'])
        #except KeyError:       
#breweryname, breweryid
```
