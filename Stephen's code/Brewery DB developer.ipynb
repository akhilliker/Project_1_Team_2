```python
#import dependencies
import pandas as pd
import numpy as np
import requests
import json
#reference api key
from config import api_key2
```

```python
#establish base url --currently using sandbox api for code exploration
base_url = "http://api.brewerydb.com/v2"
```

```python
#look at all breweries to get id and brewery names
breweryname = []
breweryid = []
breweries_query = "breweries"
breweries_url = base_url + "/" + breweries_query +"/"+ "?key=" + api_key2 + "&p=1"
breweries = requests.get(breweries_url).json()
for brewery in breweries['data']:
    breweryname.append(brewery['name'])
    breweryid.append(brewery['id'])
    print(brewery['name'], brewery['id'])
```

```python
#create dataframe of breweries and ids
breweries_df = pd.DataFrame({"Name":breweryname, "ID":breweryid})
breweries_df
```

```python
#create request url to get list of all beers for each brewery in brewery dataframe
brewery_path = "/brewery/{}/beers/"
beers_url = base_url + brewery_path + "?key=" + api_key2
```

```python
brewery = []
beername = []
beerstyle = []
for index, row in breweries_df.iterrows():
    #print(requests.get(inbev_beers_url.format(row["ID"])).json())
    try: 
        for beer in requests.get(beers_url.format(row["ID"])).json()['data']:
            try:
                brewery.append(row["ID"])
            except:
                brewery.append(None)
            try:
                beername.append(beer['name'])
            except:
                beername.append(None)
            try:    
                beerstyle.append(beer['style']['name'])
            except:
                beerstyle.append(None)
    except KeyError: 
        print(row)
        stylesdf = pd.DataFrame({"Brewery": brewery, "Beer":beername, "Style":beerstyle})        
```
