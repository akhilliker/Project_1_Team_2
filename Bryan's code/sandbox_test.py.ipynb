```python
#import dependencies
import pandas as pd
import requests
import json
import sys
import time
from sand_box import sandbox_key
```

```python
url = "https://sandbox-api.brewerydb.com/v2/"

api_key = sandbox_key

query_url = f"{url}locations/?key={api_key}"
response = requests.get(query_url).json()

print(json.dumps(response, indent=4, sort_keys=True))
```

```python
abv_list = []
id_list = []

for page_num in range(1,24):
    time.sleep(3)
    url = "https://sandbox-api.brewerydb.com/v2/beers/?key=" + api_key + "&p=" + str(page_num)
    print("Downloading", url)
    response=requests.get(url).json()
    
    for abv in response['data']:
        try:
            abv_list.append(abv['abv'])
        except:
            abv_list.append(None)
        try:
            id_list.append(abv['id'])
        except:
            id_list.append(None)
```

```python
brewery_df = pd.DataFrame({
                    "ID": id_list,
                    "Abv": abv_list
                            })
new_df = brewery_df.sort_values("Abv", ascending=False)
new_df
```
